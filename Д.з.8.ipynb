```python
import pandas as pd
```

```python
data1 = pd.read_csv("C:/Users/eldorado/Pictures/clients_data.csv")
data1.head(3)
```

```python
data2 = pd.read_csv("C:/Users/eldorado/Pictures/products_data.csv")
data2.head(3)
```

```python
data3= pd.read_csv("C:/Users/eldorado/Pictures/orders_data.csv")
data3.head(3)
```

```python
data4 = pd.read_csv("C:/Users/eldorado/Pictures/positions_data.csv")
data4.head(10)
```

```python
overall= data2.merge(data4, left_on="Id", right_on="Product_Id")
overall.head()
```

```python
overall.groupby('Name')['Number'].mean() # Вопрос4. среднее количество позиций в чеке
```

```python
overall['Sum']=overall['Cost']*overall['Number']
overall.head(50)
```

```python
tot1 = overall['Sum'].sum() # общая Заработок   без учета скидки  Вопрос 1.1. Суммарный заработок компании за все время.
tot1
```

```python
tot2 = overall['Number'].sum() # общее число заказов
tot2
```

```python
average_amount = tot1 / tot2 # Вопрос 3. средняя сумма заказа
average_amount
```

```python
over= data3.merge(overall, left_on="Id", right_on="Order_Id")
over.head()
```

```python
over['Sale_%'] = over['Sale_%']/100
over['Sale_%']
```

```python
over['Sum_']=over['Sum'] - over['Sum']*over['Sale_%'] #   Сумма с учетом скидки.
over
```

```python
over['full_address']=over['City']+over['Address']
over
```

```python
city_gr = over.groupby('Client_Id')['full_address'].unique()
city_gr = city_gr.apply(len)
city_gr = city_gr.reset_index()
city_more1 = city_gr [city_gr['full_address']>1]
city_more1 # Клиентов сделавших заказы по разным адресам.
```

```python
over.groupby(['Product_Id','City'])['Number'].sum().reset_index() # Вопрос 5 самый популярный товар в каждом городе
```

```python
over.groupby('Client_Id')['Sum_'].sum().reset_index() # Вопрос 1.2 суммарный заработок по клиентам.
```

```python
over.groupby('Year')['Sum_'].sum().reset_index() # Вопрос 1.2 суммарный заработок по годам.
```

```python
overall['Name'].value_counts() # Вопрос 2. Сколько раз был продан каждый товар.
```
